# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```sql
import pandas as pd
import matplotlib.pyplot as plt

QUERY = """
SELECT ConvertedComp
FROM master
"""

# Read the 'master' table into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Extract the 'ConvertedComp' column
converted_comp = df['ConvertedComp']

# Plot a histogram of ConvertedComp
plt.hist(converted_comp, bins=30, edgecolor='black')

# Set labels and title
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.title('Histogram of Converted Compensation')

# Display the plot
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```sql
QUERY = """
SELECT Age
FROM master
"""

# Read the 'master' table into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Extract the 'Age' column
age = df['Age']

# Create the boxplot using Pandas
age.plot(kind='box')

# Set the title
plt.title('Boxplot of Age')

# Display the plot
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```sql
QUERY = """
SELECT Age, WorkWeekHrs 
FROM master
"""

# Read the 'master' table into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Extract the 'Age' and 'WorkWeekHrs' column
age = df['Age']
work_hours = df['WorkWeekHrs']

# Create a scatter plot
plt.scatter(age, work_hours)

# Set labels and title
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.title('Scatter Plot of Age vs Work Week Hours')

# Display the plot
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```sql
QUERY = """
SELECT CodeRevHrs, WorkWeekHrs 
FROM master
"""

# Read the 'master' table into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Extract the 'Age' and 'WorkWeekHrs' column
code_hours = df['CodeRevHrs']
work_hours = df['WorkWeekHrs']

# Create a bubble plot
plt.scatter(code_hours, work_hours, s=50, alpha=0.5)

# Set labels and title
plt.xlabel('Code Rev Hours')
plt.ylabel('Work Week Hours')
plt.title('Bubble Plot of Code Rev Hours vs Work Week Hours')

# Display the plot
plt.show()
```

## Visualizing composition of data


### Pie Charts


```sql
QUERY = """
SELECT COUNT(DatabaseDesireNextYear) as Count, DatabaseDesireNextYear
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY COUNT(DatabaseDesireNextYear)
DESC LIMIT 5
"""

# Read the 'master' table into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Create a bar chart
plt.bar(df['DatabaseDesireNextYear'], df['Count'])

# Set labels and title
plt.xlabel('Database Desire Next Year')
plt.ylabel('Count')
plt.title('Bar Chart of Top 5 DatabaseDesireNextYear')

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Display the plot
plt.show()


```

Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```sql
QUERY = """
SELECT COUNT(DatabaseDesireNextYear) as Count, DatabaseDesireNextYear
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY COUNT(DatabaseDesireNextYear)
DESC LIMIT 5
"""

# Read the 'master' table into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Create a pie chart
plt.pie(df['Count'], labels=df['DatabaseDesireNextYear'], autopct='%1.1f%%')

# Set title
plt.title('Pie Chart of Top 5 DatabaseDesireNextYear')

# Display the plot
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```sql
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM Master
WHERE Age >= 30 AND Age <= 35
"""

# Read the 'master' table into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Compute the median values of WorkWeekHrs and CodeRevHrs
median_work_hours = df.groupby('Age')['WorkWeekHrs'].median()
median_code_review_hours = df.groupby('Age')['CodeRevHrs'].median()

# Set the x-axis labels
x_labels = median_work_hours.index

# Create a stacked chart
plt.bar(x_labels, median_work_hours, label='WorkWeekHrs')
plt.bar(x_labels, median_code_review_hours, bottom=median_work_hours, label='CodeRevHrs')

# Set labels and title
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs\n(Age Group: 30 to 35)')
plt.legend()

# Display the plot
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```sql
QUERY = """
SELECT ConvertedComp, Age
FROM Master
WHERE Age >= 45 AND Age <= 60
"""

# Read the 'master' table into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Compute the median values of WorkWeekHrs and CodeRevHrs
median_comp = df.groupby('Age')['ConvertedComp'].median()

# Plot the line chart
plt.plot(median_comp.index, median_comp.values, marker='o')

# Set labels and title
plt.xlabel('Age')
plt.ylabel('median_comp')
plt.title('Line Chart of Median ConvertedComp by Age (45 to 60)')

# Display the plot
plt.show()
```

```sql
QUERY = """
SELECT ConvertedComp, Age
FROM Master
WHERE Age >= 25 AND Age <= 30
"""

# Read the 'master' table into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Compute the median values of WorkWeekHrs and CodeRevHrs
median_comp = df.groupby('Age')['ConvertedComp'].median()

# Plot the line chart
plt.plot(median_comp.index, median_comp.values, marker='o')

# Set labels and title
plt.xlabel('Age')
plt.ylabel('median_comp')
plt.title('Line Chart of Median ConvertedComp by Age (25 to 30)')

# Display the plot
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```sql
QUERY = """
SELECT MainBranch, COUNT(*) AS Count 
FROM Master
GROUP BY MainBranch
ORDER BY Count
"""

# Read the 'master' table into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Set 'MainBranch' as the index
df.set_index('MainBranch', inplace=True)

# Create a horizontal bar chart
plt.barh(df.index, df['Count'])

# Set labels and title
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.title('Horizontal Bar Chart of MainBranch')

# Display the plot
plt.show()
```
